** Mostrar los datos en un *data frame* **

```python
import pandas as pd
idx= [1,2,3,4,5]
df = pd.DataFrame(d,index = idx)
df["Promedio"] = df.mean(axis= 1)
df
```

## Mostrar los nuevos datos que coincidieron con la condici칩n previa

```python
avg = ssm_1[ssm_1['Avg_Daily_Usage_Hours']>4]
avg
```

**Cambiar el nombre de una columna**

```python
ssm_1 = ssm_1.rename(columns={'Hours':'Horas', "Most_Used_Platform":'Plataforma',"Age":"Edad","Gender":"Sexo"})
ssm_1
```

**Elegir aquellos estudiantes que superan 4 horas en redes sociales**

```python
ssm_2= ssm_1[ssm_1['Horas'] > 4.0]
ssm_2
```

**Agrupar por sexo el m치ximo n칰mero de horas**

```python
groupssm_2 = ssm_2.groupby(by="Sexo")['Horas'].max()
groupssm_2
```

**Agrupar el m치ximo n칰mero de horas utilizada por plataforma**

```python
platform = ssm_2.groupby(by="Plataforma")['Horas'].max()
platform
```

Contabilizar cuantas personas utilizan determinada plataforma, se puede utilizar `count` o `size`




```python
Contar = ssm_2.groupby(by="Plataforma")['Horas'].size().sort_values(ascending=False)
Contar
```

#Visualizaci칩n de datos

Matplotlib para gr치ficar

```python
import pandas as pd
import numpy as np   # solo si vas a usar np.nan

datos = {
    "Sexo": ["F",  "M",  "F",  "M"],
    "Edad": ["18-21", "18-21", "22-25", "22-25"],
    "Value": [6.5, 8.2, 6.1, 7.5]      #  completa aqu칤 tu dato
}

df2 = pd.DataFrame(datos)
df2
```

```python
# Bibliotecas a importar
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.barplot(data=df2, x="Edad", y="Value")
plt.ylabel("Valor promedio")
plt.xlabel("Edad")
plt.tight_layout()
plt.show()
```
