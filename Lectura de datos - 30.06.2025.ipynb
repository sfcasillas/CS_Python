Importar datos **numpy** para trabajar con arrays (matrices) multidimensionales de manera eficiente, ofreciendo una amplia gama de operaciones matemáticas y lógicas sobre estos datos.

En este ejemplos se crearán valores entre 0 y 10 para cada individuo, ej: "A", "B", etc.

```python
import numpy as np

np.random.seed(1)

d= {"A": np.random.binomial(10, 0.5, 5),
    "B": np.random.binomial(10, 0.3, 5),
    "C": np.random.binomial(10, 0.6, 5),
    "D": np.random.binomial(10, 0.4, 5)
}

d
```

** Mostrar los datos en un *data frame* **
Para ver los datos, es mejor utilizar un dataframe

```python
import pandas as pd
df = pd.DataFrame(d)
df
```

Mostrar las columnas utilizando la función **columns**

```python
df.columns
```

Escoger los datos específicos de un individuo.

```python
# Escoger columna(s)
df[["B", "C"]]
```

Poner datos de forma Transpuesta

```python
#Escoger por filas y por rangos
df.iloc(1)
df.iloc[0:4:2]
```

Conocer la información de la base de datos

```python
# información sobre la BD
df.info()
```

Conocer la estadística descriptiva de los valores numéricos que se encuentren en la base de datos

```python
df.describe() #Estadistica descriptiva de los datos
```

Conocer las estadísitcas de los valores traspuestos de la base de datos

```python
df.T.describe() #Describir por renglon (Transpuesta)
```

Seleccionar una valore específico de la base de datos, proporcionando **Columna** e **Índice**

```python
print(df['A'][0])
```
